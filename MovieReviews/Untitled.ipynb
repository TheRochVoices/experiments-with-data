```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.tsv', delimiter='\t', quoting=3)
```

```python
df.sample(5)
```

```python
#cleaning the text by only keeping alphabets, converting upper case to lower case(which it already is as it 
#seems)

def genCorpus(df):
    import re
    from nltk.corpus import stopwords
    from nltk.stem.porter import PorterStemmer
    from nltk.corpus import words
    ps = PorterStemmer()
    corpus=[]
    for i in range(df.shape[0]):
        print("t-",df.shape[0]-i)
        rev = df['Phrase'][i]
        rev = rev.lower()
        rev = re.sub(pattern='[^a-zA-Z]', repl=' ', string = rev)
        rev = rev.split()
        rev = [ps.stem(word) for word in rev if not word in set(stopwords.words('english')) and word in set(words.words())]
        rev = ' '.join(rev)
        corpus.append(rev)
    return corpus
    
```

```python
corpus = genCorpus(df)
```

```python
with open("corpus.txt", "w") as f:
    for s in corpus:
        f.write(str(s) +"\n")
```
