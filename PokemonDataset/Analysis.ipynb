```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Pokemon.csv")
```

```python
df.head(10)
```

```python
df['Type 1'].value_counts()
```

```python
df['Type 1'].value_counts().plot(kind='pie', figsize=(8,8))
```

```python
water = df[df['Type 1'] == 'Water']
fire = df[df['Type 1'] == 'Fire']
grass = df[df['Type 1'] == 'Grass']
plt.scatter(water['Attack'].head(50), water['Defense'].head(50), color='Blue', label='Water')
plt.scatter(fire['Attack'].head(50), fire['Defense'].head(50),marker='*', color='Red', label='Fire')
plt.scatter(grass['Attack'].head(50), grass['Defense'].head(50),marker='+', color='Green', label='Grass')

plt.xlabel('Attack')
plt.ylabel('Defense')
plt.legend()
fig=plt.gcf()
fig.set_size_inches(12, 8)

```

Water type generally have high defense whereas Fire type have more attack. Grass type are average in both the categories.

```python
a=df.groupby(['Generation','Type 1']).count().reset_index()
a=a[['Generation','Type 1','Total']]
a=a.pivot('Generation','Type 1','Total')
a[['Water','Fire','Grass','Dragon','Normal','Rock','Flying','Electric']].plot(color=['b','r','g','#FFA500','brown','#6666ff','#001012','y'],marker='o')
fig=plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```

Grass type increased till Generation 5. Rock type remained constant. Water type showed a gradual decrease.
